## Cleaning

```python
def check_missing_columns(dfs, selected_features):
    "Check for missing columns in datasets."
    for i, df in enumerate(dfs):
        missing = [feature for feature in selected_features if feature not in df.columns]
        if missing:
            print(f"Warning: Dataset {i} missing columns: {missing}")

```

```python
def select_features(df, selected_features):
    "Select available features in a dataset."
    return df[[col for col in selected_features if col in df.columns]].copy()

def fill_missing_text_columns(df, text_columns):
    "Fill missing values in text columns"
    for col in text_columns:
        if col in df.columns:
            df[col] = df[col].astype(str).fillna("")
    return df




```

```python
def process_datasets(dfs, selected_features):
    "Process datasets by selecting features and handling missing text"
    processed = []
    text_cols = ["text", "quote/text"]
    for df in dfs:
        df_selected = select_features(df, selected_features)
        df_filled = fill_missing_text_columns(df_selected, text_cols)
        processed.append(df_filled)
    return processed
```

```python
def select_and_concat_datasets(dfs):
    "Select relevant features and concatenate datasets."
    selected_features = [
        "text", "quote/text", "author/description", "author/followers",
        "author/favouritesCount", "author/createdAt", "source", "quote/viewCount",
        "quoteCount", "twitterUrl", "type", "url", "createdAt", "author/location",
        "retweetCount", "replyCount", "likeCount"
    ]

    check_missing_columns(dfs, selected_features)
    processed_dfs = process_datasets(dfs, selected_features)
    return pd.concat(processed_dfs, ignore_index=True, sort=False) if processed_dfs else pd.DataFrame()
```

```python
# Example

wsr = select_and_concat_datasets(wsr_dfs)
```

```python
hanifa = select_and_concat_datasets(hanifa_dfs)
```

```python
morara = select_and_concat_datasets(morara_dfs)
```

```python
df = select_and_concat_datasets([hanifa, morara, wsr])
```
